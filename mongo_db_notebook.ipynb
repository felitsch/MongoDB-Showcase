## T2. Move reviews to a new collection
This code should be executed once only.

TRANSFORMATION APPLIED:

Move reviews to a new collection. Remove them from the main document.


REASONING BEHIND:

We're updating reviews very frequently. This is a problem for two reasons:

a) increased size of the documents in the listingsAndReviews_new collection, making it more "costly" do read and update;

b) frequent updates may cause contention and make it harder for Mongo to distribute the information;


EXPECTED RESULT:

- Smaller documents in the listingsAndReviews_new colection.
- A new collection just with the reviews, named REVIEWS.

COMMAND:

Note: Reviews already include the property's id.


```python
# The pipeline has three steps:
# unwind the reviews to create an individual document for each current review;
# Select the relevant columns using a projection;
# output this result directly do a new collection

pipeline = [
    {"$unwind": "$reviews"},
    {"$project": {
        "_id": "$reviews._id",
        "listing_id": "$_id",
        "comments": "$reviews.comments",
        "date": "$reviews.date",
        "reviewer_id": "$reviews.reviewer_id",
        "reviewer_name": "$reviews.reviewer_name"
    }}
    ,
    {"$out": "REVIEWS"}
]


db.listingsAndReviews_new.aggregate(pipeline)

```

## T3 - Move review scores to its own structure, applying the attribute pattern.


This should follow the attribute pattern, saving the attribute name, current average and number of comments that contributed to each customer.



#### Check some of the existing review_scores_x attributes

```python
db.listingsAndReviews_new.find_one({},{"REVIEW_SCORES":1, "review_scores_checkin":1, "review_scores_cleanliness":1, "review_scores_value":1})
```

#### Group all review_scores_* attributes inside a REVIEW_SCORES attribute. Remove the previous ones.

This aggregation creates the attribute pattern and stores it as new review_scores fields in the original documents.
The old score fields are removed from the documents using 𝑝𝑟𝑜𝑗𝑒𝑐𝑡.
𝑁𝑜𝑤 𝑤𝑒 𝑐𝑎𝑛 𝑢𝑠𝑒 group and $avg operations to retrieve the average scores per attribute.


```python
db.listingsAndReviews_new.aggregate([
  {
    "$addFields": {
      "REVIEW_SCORES": [
        {
          "CATEGORY": "CHECKIN",
          "SCORE": {"$toDouble": "$review_scores_checkin"},
          "WEIGHT": "$number_of_reviews"
        },
        {
          "CATEGORY": "CLEANLINESS",
          "SCORE": {"$toDouble": "$review_scores_cleanliness"},
          "WEIGHT": "$number_of_reviews"
        },
        {
          "CATEGORY": "COMMUNICATION",
          "SCORE": {"$toDouble": "$review_scores_communication"},
          "WEIGHT": "$number_of_reviews"
        },
        {
          "CATEGORY": "LOCATION",
          "SCORE": {"$toDouble": "$review_scores_location"},
          "WEIGHT": "$number_of_reviews"
        },
        {
          "CATEGORY": "RATING",
          "SCORE": {"$toDouble": "$review_scores_rating"},
          "WEIGHT": "$number_of_reviews"
        },
        {
          "CATEGORY": "VALUE",
          "SCORE": {"$toDouble": "$review_scores_value"},
          "WEIGHT": "$number_of_reviews"
        }
      ]
    }
  },
  {
    "$project": {
      "review_scores_checkin": 0,
      "review_scores_cleanliness": 0,
      "review_scores_communication": 0,
      "review_scores_location": 0,
      "review_scores_rating": 0,
      "review_scores_value": 0
    }
  },
  {
    "$merge": {
      "into": "listingsAndReviews_new",
      "whenMatched": "merge",
      "whenNotMatched": "insert"
    }
  }
]);
```

#### REVIEW_SCORES are all in a range 0:10, except for RATING that has a 0:100 range
#### So we divide the rating by 10 to make it consistent with the other REVIEW_SCORES

#### Update REVIEW_SCORES to store the number of reviews for each score

In order to properly update scores, we need to not only keep the current score, but also the total of reviews.
If a property has rating 4 with just one review and I rate it as 5, the score will be updated to 4.5;
if it already had 9 reviews, the update will be just to 4.1.

We initialize all of these WEIGHT values with the number of reviews. This attribute, however, is unsufficient, as we know for instance that new attributes are being added. So, for those, the next review might be the first, whilst for other attributes
there could already be dozens or hundreds of reviews behind their current score.


```python
pipeline = [
  {
    "$set": {
      "REVIEW_SCORES": {
        "$map": {
          "input": "$REVIEW_SCORES",
          "as": "review",
          "in": {
            "CATEGORY": "$$review.CATEGORY",
            "SCORE": {
              "$cond": [
                {"$eq": ["$$review.CATEGORY", "RATING"]},
                {"$divide": ["$$review.SCORE", 10]},
                "$$review.SCORE"
              ]
            },
            "WEIGHT": "$$review.WEIGHT"
          }
        }
      }
    }
  }
];

db.listingsAndReviews_new.update_many({}, pipeline);
```

### T4.  Combine scattered host-related attributes into one single attribute


TRANSFORMATION APPLIED:

Embed host-related attributes that are scattered in the document into a new attribute (called "HOST"). Then, remove these attributes from the main document.


REASONING BEHIND:

By combining all host attributes in a single field, we get host information combined in a way that's consistent and better structured. While not an attribute-pattern, this will be useful if we later want to change from linking to embedding the host data, and whenever we decide to add additional host attributes.


EXPECTED RESULT:

Less "root attributes" in the document. Better hierarchization of information.

COMMAND:


```python
# Before running this "Task":
# There's a host_id and several other attributes with the prefix host_
# There is no HOST attribute; we'll be creating it to store all other attributes
# So, after the first run, it'll be empty.

process_1 = {
    "$match": {
            "host_id": {"$exists": "true"}
        }
    }

process_2 = { "$project" : {"HOST":1, "host_id":1, "address.country":1}}

process_limit = {
        "$limit": 10
}

pipeline = [process_1, process_2, process_limit]

iterator = db.listingsAndReviews_new.aggregate(pipeline)
result = list( iterator)

pprint(result)
```

```python
## Check if attribute host_id is still present.
## For records that still have a host_id, group the attributes and move them to the HOST attribute.
## If all host_id were already removed from the root level, the first process in the pipeline will return nothing
## and so, in practice nothing will happen (in that case)
# Still kept inside the listingsAndReviews_new collection, because it's used to build the homepage and 
# it's not a very large number of attributes, whose removal would significantly shrink that collection.


process_filter_oldhost_struct = {
    "$match": {
            "host_id": {"$exists": "true"}
        }
    }


    
proc_merge = {
        "$merge": {
            "into": "listingsAndReviews_new",
            "whenMatched": [
                {
                    "$addFields": {
                        "HOST": {
                            "ABOUT": "$host_about",
                            "HAS_PROFILE_PIC": "$host_has_profile_pic",
                            "ID": "$host_id",
                            "IDENTITY_VERIFIED": "$host_identity_verified",
                            "IS_SUPERHOST": "$host_is_superhost",
                            "LISTINGS_COUNT": "$host_listings_count",
                            "LOCATION": "$host_location",
                            "NAME": "$host_name",
                            "NEIGHBOURHOOD": "$host_neighbourhood",
                            "PICTURE_URL": "$host_picture_url",
                            "RESPONSE_RATE": "$host_response_rate",
                            "RESPONSE_TIME": "$host_response_time",
                            "THUMBNAIL_URL": "$host_thumbnail_url",
                            "TOTAL_LISTINGS_COUNT": "$host_total_listings_count",
                            "URL": "$host_url",
                            "VERIFICATIONS": "$host_verifications"
                        }
                    }
                },
                {
                    "$project": {
                        "host_about": 0,
                        "host_has_profile_pic": 0,
                        "host_id": 0,
                        "host_identity_verified": 0,
                        "host_is_superhost": 0,
                        "host_listings_count": 0,
                        "host_location": 0,
                        "host_name": 0,
                        "host_neighbourhood": 0,
                        "host_picture_url": 0,
                        "host_response_rate": 0,
                        "host_response_time": 0,
                        "host_thumbnail_url": 0,
                        "host_total_listings_count": 0,
                        "host_url": 0,
                        "host_verifications": 0
                    }
                }
            ],
            "whenNotMatched": "fail"
        }
    }

    

db.listingsAndReviews_new.aggregate([
        process_filter_oldhost_struct,
        proc_merge])
    

# db.listingsAndReviews_new.find_one({}, {"HOST":1})
```

### T5. Create a new collection for storing transactions; remove the transactions from inside the properties node.

TRANSFORMATION APPLIED:

Move transactions to a new collection, keeping then inside a properties (listing) node.
We will remove this information from the listingsAndReviews_new documents, because it's not used to create the landing webpage for properties.

Updating the count and the bucket_end date for the transactions would be additional updating costs that make no sense, as we don't present any of that information on the property's home page. Also, we don't have special requirements from our customer regarding transactions. So, we just removed those attributes (transaction count, bucket start date and bucket end date).

REASONING BEHIND:

Transactions are only used as a backoffice information, we want to optimize the document used for loading property's webpages, which does not include transactions information.
By keeping transactions for the same property inside the owner's node, we ensure that opening in the backoffice will still be quite quick. 

EXPECTED RESULT:

Smaller listingsAndReviews_new pages.
Also less issues with updating documents in that collection, even though that's not something that our customers have complaines about.
No loss in performance when consulting in the backoffice the transactions for a certain property.

COMMAND:


```python
# Create the new collection 'TRANSACTIONS_BY_LISTING' with transactions grouped by 'listing_id'.


pipeline = [
    {
        "$project": {
            "listing_id": "$_id",
            "transactions": "$transactions.transactions",
        }
    },
    {
        "$group": {
            "_id": "$listing_id",
            "transactions": {"$push": "$transactions"},
        }
    },
    {
        "$project": {
            "_id": 0,
            "listing_id": "$_id",
            "transactions": 1,
        }
    },
    {
        "$out": "TRANSACTIONS_BY_LISTING"
    },
]
db.listingsAndReviews_new.aggregate(pipeline)
```

```python
# Remove the 'transactions' from the 'transactions' field in the listingsAndReviews_new collection
db.listingsAndReviews_new.update_many(
    {}, {"$unset": {"transactions": ""}}
)
```

If we want to apply the attribute pattern inside the database we can use the $addFields operation in the aggregation to create the attribute pattern and store the result as new fields in the original documents:

We can use this new structure to query the average score for an example property:

```python
# Query for the average of the evaluation metrics for all properties

result = db.listingsAndReviews_new.aggregate([
  {
    "$unwind": "$REVIEW_SCORES"
  },
  {
    "$group": {
      "_id": "$REVIEW_SCORES.CATEGORY",
      "weighted_sum": {
        "$sum": {
          "$multiply": ["$REVIEW_SCORES.SCORE", "$REVIEW_SCORES.WEIGHT"]
        }
      },
      "total_weight": {"$sum": "$REVIEW_SCORES.WEIGHT"}
    }
  },
  {
    "$project": {
      "_id": 0,
      "CATEGORY": "$_id",
      "AVG_SCORE": {"$divide": ["$weighted_sum", "$total_weight"]}
    }
  }
]);

average_scores = list(result);
print("Average ratings for all properties:");
for score_info in average_scores:
  print(score_info["CATEGORY"], ":", round(score_info["AVG_SCORE"], 2))
```

```python
def calculate_average_transaction_value(listing_id, start_date, end_date):
    # Query the database for the transactions of the given listing_id
    result = db.TRANSACTIONS_BY_LISTING.find_one({"listing_id": listing_id})

    # Extract transactions from the result
    transactions = result["transactions"][0]

    # Filter transactions within the given date range
    filtered_transactions = [transaction for transaction in transactions if start_date <= transaction["date"] <= end_date]

    # Calculate the average transaction value
    total_value = sum(float(transaction["price"]) for transaction in filtered_transactions)
    average_value = total_value / len(filtered_transactions)

    return average_value


# Example usage
listing_id = "13949781"
start_date = datetime(2016, 1, 1)
end_date = datetime(2016, 12, 31)

average_value = calculate_average_transaction_value(listing_id, start_date, end_date)
print(f"Average transaction value for listing {listing_id} between {start_date} and {end_date}: {average_value}")
```

### Question 15
Add a new review metric called x_factor to the new document and give a score of 10. Show that the average across all metrics is correctly calculated for this listing, with the query you previously developed.

```python
property_id = "110000000"

new_review_scores = {
    "CHECKIN": 10.0,
    "CLEANLINESS": 9.0,
    "COMMUNICATION": 10.0,
    "LOCATION": 5.0,
    "VALUE": 9.0,
    "X_FACTOR": 10.0
}

# Get the property document from the database
property_doc = db.listingsAndReviews_new.find_one({"_id": property_id})

# Add new categories to REVIEW_SCORES if they don't already exist
for category, new_score in new_review_scores.items():
    if not any(review["CATEGORY"] == category for review in property_doc["REVIEW_SCORES"]):
        db.listingsAndReviews_new.update_one(
            {"_id": property_id},
            {
                "$push": {
                    "REVIEW_SCORES": {
                        "CATEGORY": category,
                        "SCORE": 0,
                        "WEIGHT": 0
                    }
                }
            }
        )

# Update the scores for existing categories
for review in property_doc["REVIEW_SCORES"]:
    category = review["CATEGORY"]
    if category in new_review_scores:
        old_weight = review["WEIGHT"]
        old_score = review["SCORE"]
        new_score = new_review_scores[category]
        new_weight = old_weight + 1
        new_score_total = old_score * old_weight + new_score
        new_score_average = new_score_total / new_weight
        db.listingsAndReviews_new.update_one(
            {"_id": property_id, "REVIEW_SCORES.CATEGORY": category},
            {
                "$set": {
                    "REVIEW_SCORES.$.SCORE": new_score_average,
                    "REVIEW_SCORES.$.WEIGHT": new_weight
                }
            }
        )
```

```python
total_score_weight = 0
total_score = 0

for review in property_doc["REVIEW_SCORES"]:
    score = review["SCORE"]
    weight = review["WEIGHT"]
    total_score += score * weight
    total_score_weight += weight

if total_score_weight != 0:
    new_rating = total_score / total_score_weight
else:
    new_rating = 0

db.listingsAndReviews_new.update_one(
    {"_id": property_id},
    {"$set": {"REVIEW_SCORES.$[elem].SCORE": new_rating}},
    array_filters=[{"elem.CATEGORY": "RATING"}]
)
```
